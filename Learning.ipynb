```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("NCI60.csv")
```

```python
df.head(10)
```

```python
X=df.iloc[:,1:6831] #Virgln sonu satrlar, hepsini aldm. Virgln sa stnlar, Unnamed'i istemediim iin 1'den baladm ve 6830'a kadar istediim iin 6831'e kadar yazdm.
X
```

```python
#Scaling. Scale etmeden verirsek veriler aras fark ok byk olduu iin byk deerlerin daha nemli olduuna karar verilebilir.

sc=StandardScaler()
X_scaled=sc.fit_transform(X)
X_scaled
```

```python
#PCA - Btn girdiler iinde etkisi en byk olanlar istiyorum

pca=PCA(n_components=2) #en nemli 2 zellii istiyorum
pca_result=pca.fit_transform(X_scaled)
```

```python
print("Eigenvalues: ")
pca.explained_variance_
```

```python
print("Variances (Percentage): ")
pca.explained_variance_ratio_*100
```

```python
principalDf=pd.DataFrame(data=pca_result, columns=["PC1","PC2"]) 
```

```python
FinalDf=pd.concat([principalDf,df[["labs"]]],axis=1) #PCA iin df oluturduk, label olarak kanser trlerini ana tablodan ekledik
```

```python
#plot PCA
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(2,1,1)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("En Etkili Genler ve Hastalk")

targets=["COLON","MELANOMA","LEUKEMIA"]

for target in targets:
        indexler=FinalDf["labs"] == target #labs stunundan sadece yukarda belirlediim etiketleri getirsin
        ax.scatter(FinalDf.loc[indexler,"PC1"],
                  FinalDf.loc[indexler,"PC2"],s=50) #s=Nokta bykl
ax.legend(targets,loc="upper right") #Legend'n yeri
ax.grid()
```
