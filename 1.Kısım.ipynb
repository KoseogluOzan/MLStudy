```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df=pd.read_csv("NCI60.csv")
```

```python
df.head(10)
```

```python
X=df.iloc[:,1:6831] #Virgülün sonu satırlar, hepsini aldım. Virgülün sağı sütünlar, Unnamed'i istemediğim için 1'den başladım ve 6830'a kadar istediğim için 6831'e kadar yazdım.
X
```

```python
#Scaling. Scale etmeden verirsek veriler arası fark çok büyük olduğu için büyük değerlerin daha önemli olduğuna karar verilebilir.

sc=StandardScaler()
X_scaled=sc.fit_transform(X)
X_scaled
```

```python
#PCA - Bütün girdiler içinde etkisi en büyük olanları istiyorum

pca=PCA(n_components=2) #en önemli 2 özelliği istiyorum
pca_result=pca.fit_transform(X_scaled)
```

```python
print("Eigenvalues: ")
pca.explained_variance_
```

```python
print("Variances (Percentage): ")
pca.explained_variance_ratio_*100
```

```python
principalDf=pd.DataFrame(data=pca_result, columns=["PC1","PC2"]) 
```

```python
FinalDf=pd.concat([principalDf,df[["labs"]]],axis=1) #PCA için df oluşturduk, label olarak kanser türlerini ana tablodan ekledik
```

```python
#plot PCA
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(2,1,1)
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_title("En Etkili Genler ve Hastalık")

targets=["COLON","MELANOMA","LEUKEMIA"]

for target in targets:
        indexler=FinalDf["labs"] == target #labs sütunundan sadece yukarda belirlediğim etiketleri getirsin
        ax.scatter(FinalDf.loc[indexler,"PC1"],
                  FinalDf.loc[indexler,"PC2"],s=50) #s=Nokta büyüklüğü
ax.legend(targets,loc="upper right") #Legend'ın yeri
ax.grid()
```
