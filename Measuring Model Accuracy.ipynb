```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Advertising.csv", index_col=0)
```

```python
df.describe()
```

```python
df.head()
```

```python
X_1=df["TV"]
X_2=df["radio"]
X_3=df["newspaper"] #Input al
```

```python
type(X_1)
```

```python
X_1.shape
```

```python
y=df["sales"] #Output al
```

```python
linear_regressor = LinearRegression() #Regresyon sağlayıcı aldık
```

```python
#Dataframe'lerin şeklini düzenle

y_r=y.values.reshape(-1,1)
print(y.shape)
print(y_r.shape)
```

```python
X_1_r=X_1.values.reshape(-1,1)
X_2_r=X_2.values.reshape(-1,1)
X_3_r=X_3.values.reshape(-1,1)
```

```python
plt.scatter(X_1,y,c="orange")
plt.xlabel("TV")
plt.ylabel("Satış")
plt.title("Satış - TV")

linear_regressor.fit(X_1_r, y_r)
y_pred_1=linear_regressor.predict(X_1_r)
plt.plot(X_1,y_pred_1,color="red")

plt.show()
```

```python
plt.scatter(X_2,y,c="green")
plt.xlabel("Radio")
plt.ylabel("Satış")
plt.title("Satış - Radio")

linear_regressor.fit(X_2_r, y_r)
y_pred_1=linear_regressor.predict(X_2_r)
plt.plot(X_2,y_pred_1,color="red")

plt.show()
```

```python
plt.scatter(X_3,y,c="blue")
plt.xlabel("Newspaper")
plt.ylabel("Satış")
plt.title("Satış - Newspaper")

linear_regressor.fit(X_3_r, y_r)
y_pred_1=linear_regressor.predict(X_3_r)
plt.plot(X_3,y_pred_1,color="red")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
income = pd.read_csv("Income.csv",index_col=0)
```

```python
income.head()
```

```python
X=income["Education"]
```

```python
y=income["Income"]
```

```python
X_r=X.values.reshape(-1,1)
y_r=y.values.reshape(-1,1)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y)

lr=LinearRegression()
lr.fit(X_r,y_r)
y_pred=lr.predict(X_r)
plt.plot(X,y_pred,color="red")

plt.xlabel("Education")
plt.ylabel("Income")
plt.title("Income - Education")
plt.show()

#Machine learning, regression line ile ifade edilen "f"yi tahmin etmektir.
```

```python
#"f"yi neden tahmin ederiz?
    #tahmin yapmak
    #çıkarım yapmak
#Tahmin
    #Çoğu zaman X olur ama  y'yi tahmin etmek gerekir.
    #Kaçınılmaz hata, tahmin ederken oluşan hata payıdır
    #Amaç, kaçınılabilir hatayı azaltmaktır.
#Çıkarım
    #Tahmin kadar önemlidir, karar verici için önemlidir.
    #Çıktının tam olarak ne olacağıyla değil, girdilerin hangisinin daha önemli olduğunu anlamaya çalışmak çıkarım yapmaktır.
```

```python
#Curse of dimensionality (Boyutun laneti)
#X=30 noktasında veri yoksa, en yakın komşulara bakarız.
#Sağlık açısından tek bir komşuya bakmak mantıklı değildir.
#%10'luk bir nokta ailesine bakacak olsak, olmayan nokta etrafında verinin %10'una tekabül bir aralık alırız.
#Buna curse of dimensionality denir.
#Boyutlar arttıkça dahil etmemiz gereken noktalar arasındaki uzaklıklar artar.
```

```python
#f'yi nasıl tahmin ederiz?
#Parametrik - Nonparametrik metodlar
#Parametrikte f'nin şekliyle ilgili bir varsayımda bulunuruz.
    #Mesela f lineer olsun dersek, lineer bir model oluşur.
    #Varsayımdan sonra train dataya uyacak bir model bulmamız gerekir.
#Underfitting - Overfitting
    #Varsayım yapıldıktan sonra kendimizi kısıtlamaya başlamış oluruz.
    #Underfitting'de model, gerçek hayat kompleksitesini karşılayamaz.
    #Overfitting'de fazla karmaşıklığa yol açar.
#Nonparametrikte varsayım yapılmaz
#Fakat nonparametrik metodlar çok fazla veriye ihtiyaç duyar (Curse of dimensionality), ve maliyetlidir
```

```python
#Tahmin doğruluğu mu, Model sadeliği mi?
#Kısıtlayıcı modeller daha anlaşılırdır, esnek modeller kompleks kalırlar
#Amaç tahmin değil de çıkarım ise, anlaşılır bir model tercih edilir. Underfit olmaya yatkındırlar
#Kompleks modeller hem maliyetli hem anlaşılması zordur ve overfit olmaya yatkındırlar
```

```python
#Regresyon ve sınıflandırma
#Değişkenler sayısal (yaş, maaş) veya kategorik (Kadın, erkek) olabilir
#y sayısal ise regresyon problemi
#y kategorik ise sınıflandırma problemi
#Sıcaklığı tahmin ederken yarın kaç derece olacak sorusunun cevabı regresyon, sıcak mı soğuk mu sorusunun cevabı sınıflandırmadır.
```
