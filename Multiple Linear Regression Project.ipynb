```python
#Model tanımlama
```

```python
cars=data.copy()
```

```python
cars.head()
```

```python
#kategorik variable'ları alalım
```

```python
#kategorik sütunlar
dummies_list = ['fueltype', 'aspiration', 'carbody','drivewheel',
                'enginelocation', 'enginetype', 'cylindernumber']
```

```python
for i in dummies_list:
    temp_df=pd.get_dummies(eval("cars"+"."+i),drop_first=True)
    cars=pd.concat([cars,temp_df],axis=1)
    cars.drop([i],axis=1,inplace=True)
```

```python
cars.head()
```

```python
#dropfirst yapınca ilk değişken elenir, one-hot encoding yapıldığı için, bir değer 0 veya 1 ise diğer değerin 1 mi 0 mı oldupunu anlamış oluruz
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(cars, train_size=0.7, random_state=42)
```

```python
test_data.head()
```

```python
train_data.columns
```

```python
#scaling ((x-xmin)/(xmax - xmin))
#değişkenleri homojenleştirir, yapılmazsa model yanlış çalışır
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scale_cols = ['wheelbase', 'torque','carlength','carwidth','curbweight',
              'enginesize', 'horsepower','fueleconomy','boreratio']
```

```python
train_data[scale_cols] = scaler.fit_transform(train_data[scale_cols])
```

```python
train_data.head()
```

```python
y_train = train_data.pop('price')
```

```python
X_train = train_data
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
#RFE (Recursive Feature Elimination)
#Backward elimination'ı otomatik yapar, Her elemede en önemsiz olan gider (p değeri en büyük olan)
```

```python
rfe = RFE(lr,n_features_to_select=10)
```

```python
rfe=rfe.fit(X_train,y_train)
```

```python
rfe.support_ #değişken seçilmiş mi? T/F
```

```python
rfe.ranking_ #Seçim sırası, 1. en önemli olacak şekilde
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
X_train.columns[rfe.support_]
```

```python
X_train_rfe=X_train[X_train.columns[rfe.support_]]
```

```python
X_train_rfe
```

```python
#OLS Analizi
X_train_rfemodel=X_train_rfe.copy()
```

```python
X_train_rfemodel=sm.add_constant(X_train_rfemodel)
```

```python
lr=sm.OLS(y_train,X_train_rfemodel).fit()
```

```python
print(lr.summary())
```

```python
X_train_rfemodel=X_train_rfemodel.drop(["rotor"],axis=1)
```

```python
def train_ols(X,y):
    X=sm.add_constant(X)
    lr=sm.OLS(y,X).fit()
    print(lr.summary())
```

```python
train_ols(X_train_rfemodel,y_train)
```

```python
X_train_rfemodel=X_train_rfemodel.drop(["dohcv"],axis=1)
train_ols(X_train_rfemodel,y_train)
```

```python
X_train_rfemodel=X_train_rfemodel.drop(["five"],axis=1)
train_ols(X_train_rfemodel,y_train)
```
